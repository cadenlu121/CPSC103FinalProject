```python
### Step 2b and 2c: Building
#### Design a function to read the information and store it as data in your program
#### Design functions to analyze the data

from cs103 import *
from typing import NamedTuple, List
import csv

##################
# Data Definitions

# Goal:
# Find the percent of lifespan affecting happiness values of each country and plot that on a 
# line chart with the x-axis being ascending sorted by overall_rank, and percent on the y-axis.

CountryData = NamedTuple('CountryData', [('overall_rank', int), # in range [1, 156]
                                        ('gdp', float),  # in range [0, ...)
                                        ('social_support', float),  # in range [0, ...)
                                        ('life_expectancy', float),  # in range [0, ...)
                                        ('freedom', float),  # in range [0, ...)
                                        ('generosity', float),  # in range [0, ...)                                              
                                        ('corruption', float)]) # in range[0, ...)
# interp. the data of a Country, including the overall happiness rank (overall_rank), 
# the GDP per capita (gdp), the social support of the country (social_support), 
# the healthy life expectancy (life_expectancy), the freedom (freedom), the generosity (generosity), 
# and the perceptions of corruption (corruption)

CD0 = CountryData(147, 0.2, 0.3, 0.99, 0.1, 0.3, 1)
CD1 = CountryData(96, 0.549, 0.91, 0.331, 0.381, 0.187, 0.037)
CD2 = CountryData(9, 1.365, 1.505, 1.039, 0.584, 0.285, 0.308)
CD3 = CountryData(68, 1.183, 1.452, 0.726, 0.334, 0.082, 0.031)

@typecheck
def fn_for_country_data(cd: CountryData) -> ...:
    # template based on Compound (7 fields!!!)
    return ...(cd.overall_rank,
               cd.gdp, 
               cd.social_support, 
               cd.life_expectancy, 
               cd.freedom, 
               cd.generosity,
               cd.corruption) 


# List[CountryData]
# interp. a list of CountryData (our csv is going to be this)

LOCD0 = []                  # example for empty list
LOCD1 = [CD1]               # example for one item
LOCD2 = [CD1, CD2, CD3]     # example for multiple items

@typecheck
def fn_for_locd(locd: List[CountryData]) -> ...:
    # template based on arbitrary-sized and the reference rule 
    # description of the acc 
    acc = ... # type: ...
    
    for cd in locd:
        acc = ...(acc, fn_for_country_data(cd))
        
    return ...(acc)
```

```python
import matplotlib.pyplot as pyplot

@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename, analyzes the data, returns the result 
    """
    # Template from HtDAP, based on function composition 
    return plot_diff_vs_rank(read(filename)) 

# this is my analysis function
@typecheck
def plot_diff_vs_rank(locd: List[CountryData]) -> None:
    """
  Find the relative percent of lifespan affecting happiness ranking of each value in the csv
  and plot that on a scatterplot with the x-axis being ascending sorted by overall_rank, 
  and percent on the y-axis.
  
  We will return None (as a graph) assuming all input values lengths are the same.
  
  Since adding up each factor affecting the happiness score does not equal the total happiness score
  I used a relative percentage as a better indicator of how much lifespan influences the score compared 
  to other factors.
    """
    # return None # stub
    # Template based on visualization
    
    # set the x-axis label, y-axis label, and plot title
    pyplot.xlabel('Overall Ranking')
    pyplot.ylabel('Relative Percent of Lifetime Affecting Score')
    pyplot.title('Relative Percent of Lifetime Affecting Score vs Overall Ranking')

    # range for the axes
    # [x-min, x-max, y-min, y-max]
    # I expanded ends of max_value to see edges better
    pyplot.axis([0, max_value(locd) + 5, 0, 1])
    
    # make a scatter plot of the data
    pyplot.scatter(sorted_list_of_ranks(locd), sorted_and_difference(locd), marker='o', c='r')

    # show the plot
    pyplot.show()
    
    return None


# Template from HtDAP
@typecheck
def read(filename: str) -> List[CountryData]:
    """    
    reads information from the specified file and returns a list of country data
    """
    # return []  #stub
    # Template from HtDAP
    # loc contains the result so far
    locd = [] # type: List[Consumed]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            # you may not need to store all the rows, and you may need
            # to convert some of the strings to other types
            cd = CountryData(parse_int(row[0]), parse_float(row[4]), parse_float(row[5]), parse_float(row[6]), parse_float(row[7]), parse_float(row[8]), parse_float(row[9]))
            locd.append(cd)
    
    return locd


####################################################################################################
# These are all my helper functions I used to simplify my code.
####################################################################################################

# my x-axis values
@typecheck
def sorted_list_of_ranks(locd: List[CountryData]) -> List[int]:
    """
    return the sorted list of integers in ascending order.
    """
    # return [] # stub 
    # template based on List[CountryData]
    # used an outside function because I would have died if I didn't
    return sorted(list_of_ranks(locd))


@typecheck
def list_of_ranks(locd: List[CountryData]) -> List[int]:
    """
    return a list of the integer ranks of all the CountryData in the list of CountryData
    """
    # return [] #stub
    # template based on List[CountryData]
    
    # acc is the list of all overall ranks (as integers) in the list of country data
    acc = [] # type: List[int]
    
    for cd in locd:
        acc.append(cd.overall_rank)
        
    return acc

###################################################################################################

# my y-axis values
@typecheck
def sorted_and_difference(locd: List[CountryData]) -> List[float]: 
    """ 
    return a list of the percent difference of life expectancy affecting score after sorting the list of
    countries by score in ascending order for each country in the sheet.
    """ 
    # return [] # stub
    # template based on composition
    # this returns the list of differences after sorting by ascending order
    return list_of_differences(sort_by_ascending(locd))


@typecheck
def sort_by_ascending(locd: List[CountryData]) -> List[CountryData]:
    """
    sort the list of countries by ascending order
    and return the list of ascending order countries
    """
    # return [] # stub
    # template based on List[CountryData]
    # used an outside function because I would have died if I didn't
    
    return sorted(locd, key=lambda x: x[0])


@typecheck
def list_of_differences(locd: List[CountryData]) -> List[float]:
    """
    finds the list of percent differences in life expectancy affecting scores
    and returns the list.
    """
    # return [] # stub
    # template based on List[CountryData]
    # acc is the list of CountryData
    acc = []   # type: List[CountryData]
    for cd in locd: 
        acc.append(find_percent_difference(cd))
    return acc
    

@typecheck
def find_percent_difference(cd: CountryData) -> float:
    """
    finds the percent difference in life expectancy affecting scores
    and returns the percent difference calculated.
    """
    # return 0 # stub
    # template based on CountryData
    return (cd.life_expectancy / 
           (cd.gdp +
           cd.social_support +
           cd.life_expectancy +
           cd.freedom +
           cd.generosity +
           cd.corruption))

####################################################################################################
### These functions are for helping out the graph size
####################################################################################################

@typecheck
def max_value(locd: List[CountryData]) -> int:
    """
    return the max overall rank in the list of country data
    """
    # return 0 # stub
    # template based on composition
    return max_value_in_list(list_of_ranks(locd))


@typecheck
def max_value_in_list(loi: List[int]) -> int:
    """
    finds the max overall rank (the largest number) in the list of ranks
    
    we will use this to determine the width of our graph's x-axis.
    """
    # return 0 # stub
    # template based on List[int]
    # max_int is the largest number so far
    
    max_int = 0 # type: int
    
    for i in loi:
        if i > max_int:
            max_int = i
        
    return max_int



####################################################################################################
# tests for read / main / analyze 
####################################################################################################

# tests for read
start_testing()
expect(read("world_happiness_ranking_2019_empty.csv"), [])
expect(read("world_happiness_ranking_2019_small.csv"), [CountryData(57, 1.12, 1.402, 0.798, 0.498, 0.215, 0.06),
                                                        CountryData(72, 1.044, 1.303, 0.673, 0.416, 0.133, 0.152),
                                                        CountryData(85, 0.696, 1.111, 0.245, 0.426, 0.215, 0.041),])
expect(read("world_happiness_ranking_2019_small2.csv"),[CountryData(26, 1.159, 1.369, 0.92, 0.357, 0.187, 0.056),
                                                        CountryData(32, 1.004, 1.439, 0.802, 0.39, 0.099, 0.086),
                                                        CountryData(43, 0.985, 1.41, 0.841, 0.47, 0.099, 0.034)])
summary()


# tests for main
start_testing()
expect(main("world_happiness_ranking_2019_empty.csv"), None)
    # should return the graph with no values but with labels on all axes
expect(main("world_happiness_ranking_2019_small.csv"), None)
    # should return the graph with 3 values as a scatter plot, with labels on all axes
summary()


# tests for plot_diff_vs_rank (analysis)
start_testing()
expect(plot_diff_vs_rank(LOCD0), None)
    # should return the graph with no values but with labels on all axes
expect(plot_diff_vs_rank(LOCD2), None)
    # should return the graph with 3 values as a scatter plot, with labels on all axes
summary()


####################################################################################################
### tests for helper functions
####################################################################################################

# tests for sorted_list_of_ranks
start_testing()

expect(sorted_list_of_ranks(LOCD0), [])
expect(sorted_list_of_ranks(LOCD1), [96])
expect(sorted_list_of_ranks(LOCD2), [9, 68, 96])

summary()


# tests for list_of_ranks
start_testing()

expect(list_of_ranks(LOCD0), [])
expect(list_of_ranks(LOCD1), [96])
expect(list_of_ranks(LOCD2), [96, 9, 68])

summary()


# tests for sorted_and_difference
start_testing()

expect(sorted_and_difference(LOCD0), [])
expect(sorted_and_difference([CountryData(85, 0.696, 1.111, 0.245, 0.426, 0.215, 0.041)]),
                               [0.245 / (0.696 + 1.111 + 0.245 + 0.426 + 0.215 + 0.041)])
summary()


# tests for sort_by_ascending
start_testing()

expect(sort_by_ascending([]), [])
expect(sort_by_ascending([CountryData(85, 0.696, 1.111, 0.245, 0.426, 0.215, 0.041),
                          CountryData(57, 1.12, 1.402, 0.798, 0.498, 0.215, 0.06),
                          CountryData(72, 1.044, 1.303, 0.673, 0.416, 0.133, 0.152)]),
                 [CountryData(57, 1.12, 1.402, 0.798, 0.498, 0.215, 0.06),
                  CountryData(72, 1.044, 1.303, 0.673, 0.416, 0.133, 0.152),
                  CountryData(85, 0.696, 1.111, 0.245, 0.426, 0.215, 0.041)])
expect(sort_by_ascending([CD3, CD2, CD1]), [CD2, CD3, CD1])
summary()


# tests for list_of_differences
start_testing()

expect(list_of_differences(LOCD0), [])
expect(list_of_differences(LOCD1), [0.331 / (0.549 + 0.91 + 0.331 + 0.381 + 0.187 + 0.037)])
expect(list_of_differences(LOCD2), [(0.331 / (0.549 + 0.91 + 0.331 + 0.381 + 0.187 + 0.037)),
                                    (1.039 / (1.365 + 1.505 + 1.039 + 0.584 + 0.285 + 0.308)),
                                   (0.726 / (1.183 + 1.452 + 0.726 + 0.334 + 0.082 + 0.031))])
summary()


# tests for find_percent_difference
start_testing()

expect(find_percent_difference(CountryData(85, 0.696, 1.111, 0.245, 0.426, 0.215, 0.041)),
                              (0.245 / (0.696 + 1.111 + 0.245 + 0.426 + 0.215 + 0.041)))
expect(find_percent_difference(CountryData(57, 1.12, 1.402, 0.798, 0.498, 0.215, 0.06)),
                              (0.798 / (1.12 + 1.402 + 0.798 + 0.498 + 0.215 + 0.06)))
expect(find_percent_difference(CountryData(85, 0.696, 1.111, 0.245, 0.426, 0.215, 0.041)),
                              (0.245 / (0.696 + 1.111 + 0.245 + 0.426 + 0.215 + 0.041)))
summary()


# tests for max_value
start_testing()

expect(max_value(LOCD0), 0)
expect(max_value(LOCD1), 96)
expect(max_value(LOCD2), 96)

summary()

# tests for max_value_in_list
start_testing()

expect(max_value_in_list([]), 0)
expect(max_value_in_list([0, 1, 3, -5]), 3)

summary()
```
